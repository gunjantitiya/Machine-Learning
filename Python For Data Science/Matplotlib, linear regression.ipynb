#### ML Algorithms
* Supervised - predictive analytics
    * historical data with known target
    * machine learns patterns from known data
    * use this knowledge to predict other unknown i/p given
        * Regression - continous value in target
            * When we have continous variable as target (house price - can go as high)
                * <b>Simple linear regression</b>
                * <b>Multi linear regression</b>
        * Classification - Discrete value in target
            * When we have discreet variable as target ( fixed value - like T or F)
            * Only two possible outcome - binary classification
            * classify data in multiple class (high , med,low risk)
* Unsupervised
    * Clustering - data cluster in groups
* ReInforcement
    * Reward based learning
    * OpenAI Gym

### Linear regression
* y=mx+c
* Find optimum line closest to the points
* Find best possible m-intercept & c-coefficient closed to data points
* y= target, x = i/p features
* multi feature - multi regression mx1+c1 mx2+c2...

### Predict mpg
#### mpg is target - everything else is feature

```python
auto_data=pd.read_csv('auto-mpg.data', names=['mpg','cylinders','displacement','horsepower','weight','acceleration','modelyear','origin','carname']
                      ,index_col=False,sep="\s+")
auto_data.head()
```

```python
auto_data['horsepower_float']=auto_data['horsepower'].apply(lambda x: np.nan if x.strip() == "?" else x.strip())
```

```python
#auto_data["horsedata_float"]=auto_data["horsepower"].astype('float')
auto_data.plot('horsepower_float',"mpg", kind='line')
```

#### More horsepower - less mileage

```python
auto_data['horsepower_float']=auto_data['horsepower_float'].astype('float')
auto_data.plot('horsepower_float',"mpg", kind='scatter')
```

### Simple linear model

```python
from sklearn.linear_model import LinearRegression
```

```python
## Data cleaning - create new data with all null values rows ignored
auto_data_new=auto_data[~auto_data['horsepower_float'].isnull()]
```

```python
auto_data_new['horsepower_float'].values
```

```python
lr = LinearRegression()
## Train the model 
## Need to convert columns to ndarray column.values for fit method
## Then ndarray needs to be reshaped to -1,1 as we need 2d array, data is 1d array
lr.fit(auto_data_new['horsepower_float'].values.reshape(-1,1), auto_data_new['mpg'].values.reshape(-1,1))
#lr.fit?
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
## Formula that model came up with
Y=39.93 + ((-0.15)*X)
```

```python
## Predict mpg using this model - given horsepower value, what could be target mpg
## target mpg = 0.47 for 250 horseopower - which is bad model
## Possibly becuase lack of data and not enough data to train near perfection
## Also, this is linear model, general trend was line is decreasing - less horse power more mileage
lr.predict(np.array([250]).reshape(-1,1))
```

```python
## Let's see how far we are from original to see if model is overfitting or underfitting
## Looking at original mpg for horsepower, compute error
ypred=lr.predict(auto_data_new['horsepower_float'].values.reshape(-1,1))
ypred
```

```python
yerror = auto_data_new["mpg"].values.reshape(-1,1) - ypred
yerror

pred_values=((ypred*100)/auto_data_new["mpg"].values)
```

```python
### Smote in python for synthetic data minority oversampling
## very less data issue can be solved
```

```python
from sklearn.metrics import mean_squared_error
## mean of error
np.sqrt(mean_squared_error(auto_data_new["mpg"].values.reshape(-1,1), ypred))
```

```python
import statsmodels.api as api
```

```python
## Here we pass target first, then features to lean
## OLS is least squared regression model
## To be null hypothesis to be accepted - it should be >95%
result=api.OLS(auto_data_new['mpg'], auto_data_new['horsepower_float']).fit()
```

```python
result.summary()
```

```python
result.predict(200)
```

```python
## Plotting cooef and intercept value as line over scatterplot to see how line is formed
## We can use polynomial regression to form curve to get better accuracy

plt.scatter(auto_data_new['horsepower_float'],auto_data_new['mpg'])
plt.plot(auto_data_new['horsepower_float'], 
        ((auto_data_new['horsepower_float']*-0.15784473) + 39.93586102), color='r')
```
